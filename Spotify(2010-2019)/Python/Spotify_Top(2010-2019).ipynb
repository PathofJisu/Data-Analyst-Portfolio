```python
#Spliting data to train and test sets.
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=50)
```

```python
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
#Multinomial Logistic Regression 
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
Mul_Model = LogisticRegression(multi_class='multinomial', solver='lbfgs', random_state=50)
Mul_Model = Mul_Model.fit(x_train,y_train)
y_pred=Mul_Model.predict(x_test)
print("Accuracy:",metrics.accuracy_score(y_test,y_pred)*100)
```

```python
from sklearn.metrics import mean_squared_error, explained_variance_score
```

```python
#coefficients
print('Coefficients: {}'.format(Mul_Model.coef_))
# The mean squared error
print("Mean squared error: %.2f" % mean_squared_error(y_pred, y_test))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % explained_variance_score(y_test, y_pred))
```

```python
#Error rate 
```

```python
error = []
for i in range(1, 40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train, y_train)
    pred_i = knn.predict(x_test)
    error.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize=(12, 6))
plt.plot(range(1, 40), error, color='red', linestyle='dashed', marker='o',
         markerfacecolor='blue', markersize=10)
plt.title('Error Rate K Value')
plt.xlabel('K Value')
plt.ylabel('Mean Error')
```

```python
#When K = 13-14, the error rate closes to 0.
```

```python
from sklearn.neighbors import KNeighborsClassifier
KNN_Model = KNeighborsClassifier(n_neighbors = 13)
KNN_Model = KNN_Model.fit(x_train, y_train)
KNN_pred = KNN_Model.predict(x_test)
print("Accuracy:",metrics.accuracy_score(y_test,KNN_pred)*100)
```

```python
#Extra Trees Classifier
```

```python
from sklearn.ensemble import ExtraTreesClassifier
Xtree_Model = ExtraTreesClassifier(random_state=50)
Xtree_Model = Xtree_Model.fit(x_train, y_train)
Xtree_pred = Xtree_Model.predict(x_test)
print("Accuracy:",metrics.accuracy_score(y_test,Xtree_pred)*100)
```
